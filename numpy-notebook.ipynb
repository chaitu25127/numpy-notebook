## DOT PRODUCT 

```python

mat1=np.random.randint(10,size=(5,3))
mat2=np.random.randint(10,size=(5,3))
```

```python
mat1
```

```python
mat2
```

```python
#elementary multiplications
mat1*mat2
```

```python
# -> Dot product (n,m)(r,p) then if n==p or m==r then we can do dot product
mat3=np.dot(mat1,mat2.T)
mat3
```
